```python
from bs4 import BeautifulSoup
import requests
import numpy as np

import bs4
from selenium import webdriver

options = webdriver.FirefoxOptions()
driver = webdriver.Firefox(options=options)
url = "https://www.timberland.com/en-us/c/footwear-0100"
driver.get(url)
html = driver.page_source
soup = bs4.BeautifulSoup(html)

#data to fetch
#product_name, product_price, category, nr modelli offerti x tipologia, differenze di prezzo x tipologia
#media, minima e massima dei valori

class_product = "line-clamp-2 self-start subtitle-6" #classe nome prodotto
class_price = "gap-x-2 i-flex wrap" #classe prezzo prodotto
products = []
prices = []
for product in soup.find_all("a", class_=class_product):
    products.append(product.text)
for price in soup.find_all("span", class_=class_price):
    prices.append(price.text)

class_category = "w-full select-none px-3" #categorie {boots, shoes, etc}
categories = []
for category in soup.find_all("span", class_category):
    categories.append(category.text)

driver.quit()
```
